```python
pre_sale = pre_sale.loc[pre_sale['법정동'] == '동패동']

(ggplot(pre_sale, aes(x='아파트', y='거래금액'))
 + geom_boxplot(fill='lightblue', color='black')
 + ggtitle('파주 동패동 아파트 매매 평균거래금액')
 + theme_minimal()
 + theme(
     text=element_text(family='NanumGothic', size=10),
     axis_text_x=element_text(angle=45, hjust=1),
     axis_title=element_text(size=14)
 )
)
```

```python
import folium

# 동패동의 중심 지점 좌표 설정
center_lat, center_lon = 37.7109727276051, 126.719552504638

# 지도 생성
map = folium.Map(location=[center_lat, center_lon], zoom_start=13)

# 동패동에 있는 아파트 추출
apartments = df.loc[df['법정동'] == '동패동', '아파트']
prices = df.loc[df['법정동'] == '동패동', '거래금액']

# 아파트와 거래금액 정보를 지도에 마커로 표시
for apartment, price in zip(apartments, prices):
    folium.Marker(
        location=[center_lat, center_lon],
        popup=f"아파트: {apartment}<br>거래금액: {price}",
        icon=folium.Icon(color='blue')
    ).add_to(map)

# 지도 출력
map
```
